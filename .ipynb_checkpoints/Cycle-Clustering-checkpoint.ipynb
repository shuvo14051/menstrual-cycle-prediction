## BMI
There are lots of missing values in a form of empty string `" "`.

```python
# Replace the empty string with NaN
df['BMI'] = df['BMI'].replace(" ", np.nan)
df['BMI'] = df['BMI'].astype(float) # ValueError: could not convert string to float: ' '
```

```python
df.BMI.isnull().sum() / len(df)
```

__92% null values in BMI column__



# Replace all the `" "` values with nan

```python
df = df.replace(" ", np.nan)
```

```python
df.info()
```

## List of feature which has missing values not more than 60%

```python
null_percentages = df.isnull().mean() * 100
selected_features = null_percentages[null_percentages <= 90].index.tolist()
selected_df = df[selected_features]
selected_df.shape
```

```python
selected_df.isnull().mean()
```

```python
selected_df.info()
```

## Clustering

### Preprocessing

```python
# Remove unnecessary column
selected_df = selected_df.drop("ClientID", axis=1)
```

```python
# Change the data type from object to float
selected_df = selected_df.astype("float")
```

```python
# Fill the missing values
df_filled = selected_df.fillna(selected_df.mean())
sum(df_filled.isnull().sum())
```

```python
# Scaling
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df_filled)
```

```python
# inertia = []

# for i in range(2,8):
#     kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
#     kmeans.fit(scaled_df)
#     inertia.append(kmeans.inertia_)

# plt.plot(range(2,8), inertia)
# plt.title("The Elbow Curve")
# plt.xlabel("Number of clusters")
# plt.ylabel("Inertial values")
# plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10),
                              timings=False)
visualizer.fit(df_filled)
visualizer.show()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
kmeans.fit(df_filled)
labels = kmeans.predict(df_filled)
labels
```

```python
## Visualization
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_filled)
```

```python
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=labels, cmap='viridis')
plt.title('KMeans Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(df_filled)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(pca_result[:, 0], pca_result[:, 1], pca_result[:, 2], c=labels, cmap='viridis')
ax.set_title('KMeans Clustering')
ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')

# Add color bar
color_bar = fig.colorbar(scatter, ax=ax, label='Cluster Label')
plt.show()
```

```python
cluster_sizes = range(2, 7)
for n_clusters in cluster_sizes:
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=0)
    kmeans.fit(df_filled)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(df_filled, labels)
    print(silhouette_avg)
    
```
