# TabNet Regression

```python
df_original = pd.read_csv("data/Original_Preprocessed_data.csv")
df_original.drop("Healthy", axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = df_original.drop('LengthofCycle', axis=1)
y = df_original['LengthofCycle']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train = y_train.values
y_test = y_test.values
```

```python
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

```python
from pytorch_tabnet.tab_model import TabNetRegressor


clf = TabNetRegressor()  #TabNetRegressor()
clf.fit(
  X_train, y_train,
    max_epochs=300,
)
preds = clf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
y_true = y_test
mse = mean_squared_error(y_true, preds)
rmse = np.sqrt(mse)

print("MSE Score:", mse)
print("RMSE Score:", rmse)
```

```python
from sklearn.metrics import r2_score

r_2 = r2_score(y_true, preds)
print("R-squared Score:", r_2)
```

# Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train, y_train)

y_true =  rf_regressor.predict(X_test)
mse = mean_squared_error(y_true, preds)
rmse = np.sqrt(mse)

print("MSE Score:", mse)
print("RMSE Score:", rmse)
r_2 = r2_score(y_true, preds)
print("R-squared Score:", r_2)
```

# SVM

```python
from sklearn.svm import SVR

svr_model = SVR(kernel='rbf')  # You can choose different kernels like 'linear', 'poly', 'rbf', etc.

svr_model.fit(X_train, y_train)

y_pred = svr_model.predict(X_test)

# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
r_2 = r2_score(y_true, y_pred)
print("R-squared Score:", r_2)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression()

linear_reg.fit(X_train, y_train)

y_pred = linear_reg.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
